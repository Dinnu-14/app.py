```python
import pandas as pd
```

```python
data = pd.read_csv('c:/dinnu/loan_data.csv')
```

```python
data
```

```python
data
```

```python
data['Gender']=data['Gender'].map({'Male':0,'Female':1})
data['Married']=data['Married'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['Loan_Status'], axis=1)
y=data['Loan_Status']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Model=DecisionTreeClassifier()
Model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Predictions= Model.predict(X_test)
accuracy_score(Predictions,y_test)
```

```python
from sklearn.tree import plot_tree,export_text
```

```python
tree_rules=export_text(Model,feature_names=['Gender','Married','AI','LM'])
plot_tree(Model,feature_names=['G','M','AI','LM'],class_names=['y','n'],filled=True,)
```

```python
import pandas as pd
```

```python
data = pd.read_csv('c:/dinnu/loan_data.csv')
```

```python
data
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)	
```

```python
data
```

```python
data['Gender']=data['Gender'].map({'Male':0,'Female':1})
data['Married']=data['Married'].map({'Yes':1,'No':0})
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['Loan_Status'],axis=1)
y=data['Loan_Status']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Model=DecisionTreeClassifier()
Model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Predication=Model.predict(X_test)
accuracy_score(Predication,y_test)
```

```python
import joblib
joblib.dump(Model,'c:/dinnu/loan_data.joblib')
```

```python
%%writefile app.py
import streamlit as st
import joblib

clf = joblib.load('c:/dinnu/loan_data.joblib')


st.title('LOAN APP BY DINESWAR')
g=st.number_input('Enter Gender 0:Male 1:Female')
m=st.number_input('Enter Martial status 0:No 1:Yes')
ai=st.number_input('Enter Loan amout in thousands')
la=st.number_input('Enter Loan amount thousands')

if st.button('PREDICT'):
    Prediction = clf.predict([[g,m,ai,la]])
    if Prediction == 'Y':
        st.text('LOAN IS APPROVED')
    else:
        st.text('LOAN IS REJUCTED')
```
